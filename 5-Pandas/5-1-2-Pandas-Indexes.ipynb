## Indexes



In lists and dictionary, the keys are unique. So, in arrays, lists and dictionaries, we have this uniqueness in the mapping of values.

So, if we have a specific index value, we will get only one possible value to lookup.

But that's not necessarily what an index gives us here (in case of Pandas). A key is sometig that uniquely identifies a single element of the collection. But an index is ot that. An index is a mapping but it doesn't have to be unique and we will come back to that.

So, let's look at **sequence types**.



Now, in the last lines above, with pandas, we can define an explicit index in addition to the implicit index.

So, the implicit index, which is the positional index, always exists. But we can define additionally another index on the sequence, and that's called an explicit index because we are defining it explicitly.

So, for example, we may decide that we want this particular index consisting of these strings: `first, second, third, fourth` and we are going to associate the elements of the index with the elements in the list. So again, this is an associative array.

Previously, we've seen that dictionaries were a type of associative array. This is now another type of associative array that we are going to create in pandas using these indexes.

### Pandas Indexes

The datatype for the Pandas indexes is the index data type and it is the most generic type of index.

So indexes contain elements and indexes are based o numpy arrays and so they themselves had an implicit positional index.

Now, let's see what ther types of indexes can be created in Pandas:

```Python
# Float64Index
idx = pd.Index([1, 2.5])

# Index (object)
idx = pd.Index(['element 1', 'element 2'])

# Int64Index
idx = pd.Index([2, 4, 6, 8, 10])

## Some common operations on Index:

# Array like behavior: let's check what's at index 0
idx[0]

# Let's slice it:
idx[1,4]

# reverse it by using extended slicing
idx[::-1]

# use fancy indexing
idx[[1,3,4]]

# let's do boolean masking
idx = pd.Index(['Delhi', 'Kolkata', 'Mumbai', 'Indore', 'Mandsaur']
idx[idx != 'Kolkata']
```
