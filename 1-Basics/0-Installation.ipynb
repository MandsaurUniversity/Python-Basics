# How to prepare your computer to be Python ready

You can prepare your computer to use Python and Jupyter Notebook in either of the two ways:
1. [Through __Comand Prompt__ (a little complex)](#Through-Command-Prompt)
2. [Using __Anaconda__ application (simpler way)](#Alternate-way-to-install-Python-and-Jupyter-Notebook)

#### Notes:
- We require _**Python**_ and _**Jupyter Notebook**_ installed on our computer before proceeding further.
- To install these two softwares, we can follow either one of the two methods as mentioned below.
- _**Jupyter Notebook**_ is an easy to use development environment and navigator launcher tool for _**Python**_.
- _**Anaconda**_ software helps us install _**Jupyter Notebook**_ along with the _**Anaconda**_ distribution of _**Python**_.

## Through Command Prompt

### Install Python on Windows Computer

1. Go to the official Python website (URL: https://www.python.org/downloads/windows/)


2. Under __Stable Releases__, download __Windows installer (64-bit)__

> 


3. While installing, ensure that you click the checkbox: __Add Python to environment variables__

> 

### Ensuring whether environment variables are set correctly or not

1. Open __command prompt__ (press _**Windows key + R**_ and type _**cmd**_ and press _**Enter key**_ or click **OK** button)

> 

2. Enter this command:

> python --version

3. If you see a version (ex: Python 3.11.4) then you have successfully installed Python.

> 

### Install Jupyter Notebook

1. Open **command prompt** again and enter this command:

> python -m pip install --upgrade pip

> 

2. After successful execution of the above command, enter this command:

> python -m install jupyter

> 


## Alternate way to install Python and Jupyter Notebook

1. Download __Anaconda__ from this page: https://www.anaconda.com/download

> 

2. Install Anaconda and open it. Click on the Launch button under __Jupyter Notebook__ tab

>
