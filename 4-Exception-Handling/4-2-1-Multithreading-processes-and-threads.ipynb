# Multithreading

## motivation behind multi-threading

- By default, programming languages are sequential, which means that they execute the statements and commands on a line by line basis.

For example: consider the below program written in Java.
```Java
public static void main(String[] args) {
    initializeArrays();
    downloadStocks();
    initializeTimeSeriesModels();
    makePredictions();
}
```

Consider this:
- In this, we have the main method
- It calls initializeArrays(), downloadStops(), initializeTimeSeriesModels() and makePredictions() in sequence.
- Java is going to execute the first command, which means that initializing the arrays.
- After finishing the execution of this method, Java is going to consider the next line, and it starts downloading stocks from the web.
- Then it starts initializing time series models.
- And finally make predictions.

So Java, as a programming language, is sequential in a single threaded application. These methods will be called one by one, and we have to wait for them to finish one by one.

So, for example:
- the download stocks method may take two or three minutes to finish.
- Then, we have to wait for that method to finish in order to start initializing the time series related models.
This is not the best solution possible because time consuming operations such as downloading the data from the web may freeze the application and the users may not know what's happening.

**So, this is the first motivation behind multithreading.**

Other relevant reason for multithreading could be separating one or more tasks which are time consuming and might be subject to interference by the execution of order tasks.

- So, for example, if we are dealing with a stock market related software that's able to download data from the web, for example, with the half of the Yahoo Finance API
- It may take several minutes to fetch the data, but we want to make sure that the application is not frozen.
- The solution is that we have to create a distinct Strad for the download operation. And during this procedure, the user can do whatever he or she wants in the application.

A good example is copying directories. For example, I'm going to copy a huge directory, and as you can see during this procedure, the operating system doesn't freeze. So this is exactly why multithreading is extremely powerful, because here this operation, as far as copying all the items are concerned, are handled by a different thread, which means that it is totally independent from the main Thread.

### Definition:

#### Process
A process is an instance of a program execution.
- When you open a software or a web browser, these are going to be distinct processes.
- The operating system assigns distinct registers, stack memory and heap memory to every single process.
- So every single process has its own registers, program, counter, stack, memory, heap memory.

#### Threads
A thread is a lightweight process.
- It is a unit of execution within a given process (which means that a given process may have several threads)
- Each thread in the process shares the memory and resources and this is why programmers have to deal with concurrent programming and multithreading.

#### Multithreading
Multithreading is the ability of the central processing unit to execute multiple processes or strands. Concurrently, both processes and threads are independent sequences of execution.
- Creating a new thread requires fewer resources than creating a new process.
- This is why multi-threading is usually dealing with its threads and not processes.
