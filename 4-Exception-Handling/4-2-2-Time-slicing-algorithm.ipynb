# Time-slicing algorithm

Let's assume that we have just a single processor and we have case threads, so we have more than one thread in our application.
- In the previous Jupyter notebook, we learnt that a given process may contain several threads.
- In that case, we have case threads, and the problem is that the single processor has to deal with these threads.
- The solution is the so-called time slicing algorithm, so processing time for a single core is shared among processes and threads.

**This is called _time slicing._**

Let's talk about a concrete example when we have two threads: 
- thread #1 and
- thread #2

Because we have just a single process or core, it has to be shared among the given threads.




So the CPU will be there on the first thread for a short amount of time, then on the second thread for a short amount of time.

Then the first thread again, then the second thread and so on.

- So this is how a single processor is able to handle multiple threads for a short amount of time.
- It's going to run the first thread, then the second strand and so on.
- So this is why it is called time slicing

Because as you can see, the processor is going to assign a very, very small time slices or short time slots for the given threads.

OK, so this is the time slicing algorithm, and this is how a single processor core is able to run multiple threads.

- There's a huge difference between multithreading and parallel computing.
- For parallel computing, every single processor core is going to have just a single strand, which means that they can run the thread continuously so there is no need for time slicing anyways.

In the following notebook, we'll talk about parallel computing, and it's going to be clear what is the exact difference between multithreading and parallel programming in this lecture. As per this time slicing algorithm, a single processor core is able to handle multiple threads.

The processor is going to handle every single thread for a short time period.
